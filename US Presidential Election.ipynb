### Feature Extraction

```python
# Calculate the percentages of votes in final election
def calculate_pct(col):
  return (final[col]/final['Popular Vote (Total)']) * 100

final['Popular Vote Pct (Trump)'] = calculate_pct('Popular Vote (Trump)')
final['Popular Vote Pct (Clinton)'] = calculate_pct('Popular Vote (Clinton)')
final['Popular Vote Pct (Others)'] = calculate_pct('Popular Vote (Others)')
```

```python
# Add the total number of electors
total_electors = []
for i,j in final.iterrows():
  total_electors.append(j['Electoral Vote (Trump)'] + j['Electoral Vote (Clinton)'])

final['Elector Number'] = total_electors
```

```python
# Winner of the state
winner = []
for i,j in final.iterrows():
  if j['Electoral Vote (Trump)'] > j['Electoral Vote (Clinton)']:
    winner.append('Republican')
  else:
    winner.append('Democrat')

final['Winner'] = winner
```

The Electoral College is a group of electors who represent the final vote of each state. Nationwide there are 538 electors, and a candidate needs 270 votes to win the race. 

All states except Maine and Nebraska award all of their votes to the candidate who achieved the majority in the popular vote of the state.

Electoral College is a criticized concept, as it represents smaller states disproportionally, but ultimately decides the result of the election. So, how did the electoral college vote in 2016?

### Swing States

In some states’ voting patterns, both sides received practically the half of the votes. These states shift between parties depending on the political climate. Commonly, these states are called swing states, and they are the prime target for both parties. We will find out which states do not lead a dramatic difference in their voting share, and discover which swing states accommodates the largest groups of population.

```python
# Population by state
state_names = list(states['NAME'].unique())
state_info = county['area_name'].str.contains("|".join(state_names))
state_info = county[state_info]
county_mask = state_info['area_name'].str.contains('County')
state_info = state_info[~county_mask]
```

```python
# Add winner information
state_info = state_info.merge(final, how='left',
                              left_on='state_abbreviation', right_on='State')
state_info = state_info[['state_abbreviation', 'area_name',
                         'Population, 2014 estimate', 'Winner']]
state_by_pops = state_info.sort_values('Population, 2014 estimate',
                                       ascending=False).head(10)
```

```python
# Plot the largest states
sns.set_style('whitegrid')
fig, ax = plt.subplots(figsize=(10, 8))
sns.barplot(x='area_name',y='Population, 2014 estimate',
       hue='Winner', data=state_by_pops, dodge=False, palette='seismic')
plt.bar_label(ax.containers[0], fmt ='%d', rotation=15)
plt.bar_label(ax.containers[1], fmt ='%d', rotation=15)
plt.xticks(rotation = 30)
ax.ticklabel_format(style='plain', axis='y')
plt.title('States by Population and Political Party', fontsize=20)
plt.xlabel('States', fontsize=14)
plt.ylabel('Population', fontsize=14)
#ax.legend(graph, handles=[color='b', color='r'], ['Republican', 'Democrat'])

plt.show()
```

As we see here, Clinton could win only 3 states out of the 10 most populated states. In order for Democrats to win the next election, the Democratic Party needs to raise this number. Let’s focus on the easier targets. Which one of these highly populated states are swing states?

```python
# Mask swing states
swing_mask = (final['Popular Vote Pct (Clinton)'] >= 45) & (final['Popular Vote Pct (Clinton)'] < 50)
swing_states = final[swing_mask]
```

```python
# Merge swing states with population information
swing_states = swing_states.merge(state_info, how='left', left_on='State',
                   right_on='state_abbreviation')
# Sort by population
top_swing_states = swing_states.sort_values('Population, 2014 estimate',
                                            ascending=False).head(10)
```

```python
# Color information
colors = []
for i,j in top_swing_states['Winner_x'].iteritems():
  if j == 'Republican':
    colors.append('#ea6a6a')
  else:
    colors.append('#6a6aea')
```

```python
# Instantiate axes
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 8))
# Custom colors
red = mpatches.Patch(color='#ea6a6a', label='Republican Win')
blue = mpatches.Patch(color='#6a6aea', label='Democrat Win')
# Visualize charts
top_swing_states.plot.bar('area_name','Popular Vote Pct (Clinton)', ax=ax1,
                          color=colors, title='Democratic Share', fontsize=10)
top_swing_states.plot.bar('area_name','Popular Vote Pct (Trump)', ax=ax2,
                          color=colors, title='Republican Share')
# Label bars
ax1.bar_label(ax1.containers[0], fmt = '%.2f')
ax2.bar_label(ax2.containers[0], fmt = '%.2f')
# Main title
plt.suptitle('Swing States by Vote Percentage and Political Party', fontsize=20)
# Custom legend
ax1.legend(handles=[red, blue], loc=4, facecolor='white', framealpha=1)
ax2.legend(handles=[red, blue], loc=4, facecolor='white', framealpha=1)
# Set titles
ax1.set_xlabel('States (Sorted by Population)', fontsize=14)
ax2.set_xlabel('States (Sorted by Population)', fontsize=14)
ax1.set_ylabel('Vote Percentage', fontsize=14)
plt.show()
```

### Rural Counties of Pennslvania

Democrats are already doing a great job in metropolitan counties. In order to flip the swing states we have checked so far, they also need to convince some of the rural areas. In order to do so, we should first understand what a rural county is and how its demographics differ from urban counties.

According to the US Census Bureau, any areas with a population larger than 50,000 are considered urban. Although the Census Bureau does not define ‘rural’, we could use the urban definition to find out which areas are not urban, and therefore rural. [[2]](https://www.census.gov/content/dam/Census/library/publications/2019/acs/ACS_rural_handbook_2019_ch01.pdf)

Let’s look back at Pennsylvania, this time focusing on provincial counties. By doing this, we aim to achieve to recognize the demographic differences between rural and urban areas. The results from this section will help us see which demographics the Democratic Party cannot reach.

```python
# Merge counties with final votes
only_county_final = county_final.dropna()
only_county_final['fips'] = only_county_final['fips'].astype('int')
counties_final = county.merge(only_county_final, how='left', on='fips')
```

```python
# Mask candidates
counties_trump = counties_final[counties_final['cand'] == 'Donald Trump']
counties_clinton = counties_final[counties_final['cand'] == 'Hillary Clinton']
```

```python
# Function to mask rural and urban counties
def rural_urban(state_str, urban_areas):
  state_final = county_final[county_final['st'] == state_str]
  state_final = state_final[state_final['cand'] == 'Hillary Clinton']
  rural_mask = (county['state_abbreviation'] == state_str) & (county['Population, 2014 estimate'] < 50000)
  rural_area = county[rural_mask]
  rural_area = rural_area.merge(state_final, how='left',
                                left_on='area_name', right_on='county')
  urban = county[county['area_name'].str.contains("|".join(urban_areas))]
  urban = urban[urban['state_abbreviation'] == state_str]
  urban = urban.merge(state_final, how='left',
                                left_on='area_name', right_on='county')
  
  return rural_area, urban

# Pennsylvania urban vs rural
penn_urban_names = list(county[county['state_abbreviation'] == 'PA']
        .sort_values('Population, 2014 estimate', ascending=False)[1:6]['area_name'])
rural_penn, urban_penn = rural_urban('PA', penn_urban_names)
```

```python
# Function to visualize rural and urban lead votes
def lead_party(rural, urban):
  rural_lead = []
  for i,j in rural['lead'].iteritems():
    if j == 'Donald Trump':
      rural_lead.append('#ea6a6a')
    else:
      rural_lead.append('#6a6aea')
  urban_lead = []
  for i,j in urban['lead'].iteritems():
    if j == 'Donald Trump':
      urban_lead.append('#ea6a6a')
    else:
      urban_lead.append('#6a6aea')
  sns.set_style('whitegrid')
  fig, (ax1, ax2) = plt.subplots(1,2, figsize=(20,9))
  rural.plot.bar(x='area_name', y="Population, 2014 estimate", ax=ax1,
                 color=rural_lead, legend=False, title='Rural Counties')
  urban.plot.bar(x='area_name', y="Population, 2014 estimate", ax=ax2, 
                 color=urban_lead, legend=False, title='Urban Counties')
  plt.suptitle('Population of Rural and Urban Areas', fontsize=20)
  ax1.set_ylabel('Population')
  ax1.set_xlabel('Counties')
  ax2.set_ylabel('Population')
  ax2.set_xlabel('Counties')
  ax1.bar_label(ax1.containers[0], fmt = '%d', rotation=90, padding=5)
  ax2.bar_label(ax2.containers[0], fmt = '%d')

# Pennsylvania rural urban population and lead party
lead_party(rural_penn, urban_penn)
```

On the left side, we see the counties which are identified as rural. On the right side, the five largest counties in the Pennsylvania, are plotted. Allegheny County is the place where the second largest city of Pennsylvania is located: Pittsburgh. Other four counties make up the Philadelphia Metropolitan Region. Low population counties were all claimed by the Republicans and the Democrats won the large urban areas.

### Primaries

Primary elections play a huge role in the candidate selection process of each party. Before the final elections, each state held primary elections to find the candidate of each party. The candidate receiving the majority of the popular vote earns the candidacy. 

In the previous sections, we have concluded that the rural areas decided the fate of the election. And we have discovered the differences between rural areas and urban areas. In this part of the analysis, we will look at how each candidate performed in primaries, with a focus on rural areas. We will see if a different candidate would perform better in rural counties.

```python
# Function to merge urban rural data to primaries
def primary_urban_rural(rural, urban):
  rural_primary = rural.merge(primary, how='left', left_on='fips_x',
                                    right_on='fips')
  rural_primary = rural_primary\
  .drop(rural_primary[rural_primary['party'] == 'Republican'].index)
  urban_primary = urban.merge(primary, how='left', left_on='fips_x',
                                    right_on='fips')
  urban_primary = urban_primary\
  .drop(urban_primary[urban_primary['party'] == 'Republican'].index)
  return rural_primary, urban_primary
```

```python
# Urban rural primary data Pennsylvania
rural_penn_primary, urban_penn_primary = primary_urban_rural(rural_penn, urban_penn)
```

```python
# Function to visualize rural primary data
def rural_primary_plot(rural_data, state_str):
  fig, (ax1, ax2) = plt.subplots(1,2, figsize=(20, 8), sharey=True)
  sns.barplot(x='area_name', y='fraction_votes', data=rural_data,
            hue='candidate', ax=ax1)
  sns.barplot(x='candidate', y='fraction_votes', data=rural_data, ax=ax2)
  plt.suptitle(f'Candidates by Rural Vote Share in Primaries ({state_str})',
             fontsize=20)
  ax1.set_xticklabels(ax1.get_xticklabels(), rotation=90)
  ax1.set_ylabel('Vote Share (Percentage)')
  ax1.set_xlabel('Rural Counties')
  ax2.set_xlabel('Candidates')
  ax2.set_ylabel('')
  ax1.bar_label(ax1.containers[0], fmt = '%.2f', rotation=90, padding=5)
  ax1.bar_label(ax1.containers[1], fmt = '%.2f', rotation=90, padding=5)
  ax2.bar_label(ax2.containers[0], fmt = '%.2f', padding=10, fontsize=15)
  sns.move_legend(ax1, "center right")
  plt.show()

# Pennsylvania rural primary graph
rural_primary_plot(rural_penn_primary, 'Pennsylvania')
```

The rural counties of Pennsylvania were more in favor of Bernie Sanders than Hillary Clinton in the Democratic Party Primaries. Perhaps, the rural communities found the ideas of Sanders closer to their needs than Hillary Clinton's.
